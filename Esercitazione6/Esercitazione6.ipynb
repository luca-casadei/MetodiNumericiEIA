**Sperimentazione numerica**

```python
import numpy as np
import matplotlib.pyplot as plotter
#Definizione delle funzioni
f = lambda x: np.exp(-x) - (x + 1)
a = -1.0
b = 2.0
alpha = 0.0
range = np.linspace(a,b,200)
plotter.title("Funzione f")
plotter.xlabel("x")
plotter.ylabel("y")
plotter.plot(range,f(range),"b-")
plotter.show()
f2 = lambda x: np.log2(x + 3.0) - 2.0
a2 = a
b2 = b
alpha2 = 1.0
range2 = np.linspace(a2,b2,200)
plotter.title("Funzione f2")
plotter.plot(range2,f2(range2),"b-")
plotter.xlabel("x")
plotter.ylabel("y")
plotter.show()
f3 = lambda x: np.sqrt(x) - (x**2.0) / 4.0
a3 = 1.0
b3 = 3.0
alpha3 = 2.0**(4.0/3.0)
range3 = np.linspace(a3,b3,200)
plotter.title("Funzione f3")
plotter.xlabel("x")
plotter.ylabel("y")
plotter.plot(range3,f3(range3),"b-")
plotter.show()
```

```python
#Metodo di bisezione
import zeros
zero,iterazioni,centri = zeros.bisezione(f,a,b,1e-12)
print("Metodo di bisezione:")
print("Zero:",zero)
print("Iterazioni:",iterazioni)
err_bis = np.abs(np.array(centri) - alpha)
plotter.title("Errore bisezione")
plotter.semilogy(np.arange(iterazioni),err_bis,"r--o")
plotter.show()
```
